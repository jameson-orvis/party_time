```python
G_sub = nx.subgraph(G, labels[labels['scene_label'] == 'Dark Trap']['track_owner'])
```

```python
threshold = 30
edges_above_threshold = [(u, v) for u, v, d in G_sub.edges(data=True) if d['weight'] > threshold]
G_sub_again = G_sub.edge_subgraph(edges_above_threshold)
```

```python
G_sub_again.number_of_nodes()
```

```python
degree_threshold = 20
nodes_above_threshold = [node for node, degree in dict(G_sub_again.degree()).items() if degree >= degree_threshold]
```

```python

G_sub_final = G_sub_again.subgraph(nodes_above_threshold)
G_sub_final.number_of_nodes()
```

```python
users = list(set(list(G_sub_final.nodes())))
```

```python
data = {
    'users' : users,
    'index' : list(range(len(users)))
}
df = pd.DataFrame(data)
```
